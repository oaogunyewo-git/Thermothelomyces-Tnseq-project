```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the Barseq pool count file into a DataFrame
file_path = '/usr2/people/shollyt22/shollyt22/TnSeq_BarSeq_sequencings/Sequencing_with_the_13k_mutants/for_analysis/TNSeq_13k_mutants_mapping_output/TNSeq_barseq_outputs_using_combined_tnseq_poolfile_all_temp_05_23_2024/poolCount.txt'
df = pd.read_csv(file_path, sep='\t')

# Ensure all chromosome values are strings
df['scaffold'] = df['scaffold'].astype(str)
df['chromosome'] = df['scaffold'].str.extract(r'(chr\d+)', expand=False)

# Check that the 'position' column is correct and ensure it's numeric
df['position'] = pd.to_numeric(df['pos'], errors='coerce')  # Adjust to the correct column name for genomic position
df.dropna(subset=['position'], inplace=True)  # Remove rows with invalid 'position'

# Chromosome sizes dictionary
chromosome_sizes = {
    'chr1': 10931058,
    'chr2': 5454242,
    'chr3': 5062665,
    'chr4': 4710208,
    'chr5': 4342007,
    'chr6': 4127677,
    'chr7': 4116359,
}

# Define bin size (10,000 base pairs)
bin_size = 50000  # Set bin size to desired value

# Create bins for the position within each chromosome
df['bin'] = (df['position'] // bin_size) * bin_size

# Ensure the chromosome column is a string
df['chromosome'] = df['chromosome'].astype(str)

# Group by chromosome and bin to count inserts
inserts_per_chromosome_bin = df.groupby(['chromosome', 'bin'])['barcode'].count().reset_index()

# Function to extract the numerical part of the chromosome for sorting
def extract_numeric_chromosome(chromosome):
    if isinstance(chromosome, str):
        return int(chromosome.replace("chr", ""))
    else:
        return chromosome  # Handle unexpected types gracefully

# Sort chromosomes in correct numerical order
inserts_per_chromosome_bin = inserts_per_chromosome_bin.sort_values(
    by=['chromosome', 'bin'],
    key=lambda x: x.astype(str).map(extract_numeric_chromosome),
    ignore_index=True
)

# Plot the bar plot with bins representing chromosomes and sub-bins based on genomic position
plt.figure(figsize=(15, 8))

# Plot chromosomes sequentially with their respective bin sizes
chromosomes = sorted(chromosome_sizes.keys(), key=extract_numeric_chromosome)
current_offset = 0
bin_width = bin_size * 0.8

chromosome_midpoints = []  # List to store the midpoints for x-ticks

for chromosome in chromosomes:
    if chromosome in inserts_per_chromosome_bin['chromosome'].unique():
        chromosome_data = inserts_per_chromosome_bin[inserts_per_chromosome_bin['chromosome'] == chromosome]
        plt.bar(
            chromosome_data['bin'] + current_offset, 
            chromosome_data['barcode'], 
            width=bin_width, 
            label=chromosome
        )
        # Calculate the midpoint of the current chromosome for x-tick label
        chromosome_midpoint = current_offset + chromosome_sizes[chromosome] / 2
        chromosome_midpoints.append(chromosome_midpoint)
        current_offset += chromosome_sizes[chromosome]

# Add labels and titles with increased font size
plt.title('Total Inserts per Chromosome by Genomic Bins', fontsize=16)
plt.xlabel('Chromosome Position', fontsize=14)
plt.ylabel('Total Inserts', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)

# Set the x-ticks to be at the midpoints of each chromosome with increased font size
plt.xticks(chromosome_midpoints, chromosomes, rotation=45, ha='center', fontsize=12)
plt.yticks(fontsize=12)  # Increase the font size of y-ticks

plt.legend(title='Chromosomes', fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```

# To do statistical t-test

### To handle rows where some values are zero and convert them to NaN before performing the Mann-Whitney U test

```python
import pandas as pd
import numpy as np
from scipy.stats import mannwhitneyu

# Load the dataset
df = pd.read_csv('/usr2/people/shollyt22/shollyt22/TnSeq_BarSeq_sequencings/Sequencing_with_the_13k_mutants/for_analysis/TNSeq_13k_mutants_mapping_output/TNSeq_barseq_outputs_using_combined_tnseq_poolfile_all_temp_05_23_2024/normalized_barseq_pool_counts_05_28_2024.csv')

# Convert all columns to numeric, coercing errors to NaN
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Replace zeros with NaN
df.replace(0, np.nan, inplace=True)

# Define the data columns for the existing and new conditions
data_37C = df[['KFRB007_1', 'KFRB008_1', 'KFRB019_1', 'KFRB020_1']]
data_45C = df[['KFRB021_1', 'KFRB022_1', 'KFRB023_1', 'KFRB024_1']]
data_50C = df[['KFRB025_1', 'KFRB026_1', 'KFRB027_1', 'KFRB028_1']]



# List to store p-values for existing conditions
p_values_37_vs_45 = []
p_values_37_vs_50 = []

# Iterate over rows and perform Mann-Whitney U test for each
for index, row in df.iterrows():
    # Drop NaN values from the datasets
    data_37C_values = row[data_37C.columns].dropna()
    data_45C_values = row[data_45C.columns].dropna()
    
    # Perform the Mann-Whitney U test for 37C vs. 45C
    if len(data_37C_values) > 0 and len(data_45C_values) > 0:
        statistic, p_value = mannwhitneyu(data_37C_values, data_45C_values, alternative='two-sided')
        p_values_37_vs_45.append(p_value)
    else:
        p_values_37_vs_45.append(np.nan)

    # Perform the Mann-Whitney U test for 37C vs. 50C
    #data_37C_values = row[data_25C.columns].dropna()
    data_50C_values = row[data_50C.columns].dropna()

    if len(data_37C_values) > 0 and len(data_50C_values) > 0:
        statistic, p_value = mannwhitneyu(data_37C_values, data_50C_values, alternative='two-sided')
        p_values_37_vs_50.append(p_value)
    else:
        p_values_37_vs_50.append(np.nan)

# Add the p-values to the DataFrame
df['MannWhitneyU_p_value_37_vs_45'] = p_values_37_vs_45
df['MannWhitneyU_p_value_37_vs_50'] = p_values_37_vs_50

# Save the DataFrame with the calculated p-values to a new CSV file
output_path = '/usr2/people/shollyt22/shollyt22/TnSeq_BarSeq_sequencings/Sequencing_with_the_13k_mutants/for_analysis/TNSeq_13k_mutants_mapping_output/TNSeq_barseq_outputs_using_combined_tnseq_poolfile_all_temp_05_23_2024/output_with_mannwhitneyu_p_values_zero_values_removed_combined.csv'
df.to_csv(output_path, index=False)
```
