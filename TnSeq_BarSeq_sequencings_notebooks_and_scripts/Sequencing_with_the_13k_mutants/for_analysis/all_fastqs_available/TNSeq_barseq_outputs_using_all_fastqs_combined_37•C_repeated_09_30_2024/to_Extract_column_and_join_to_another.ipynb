```python
import pandas as pd

# Load the files
file1 = pd.read_csv('/usr2/people/shollyt22/shollyt22/TnSeq_BarSeq_sequencings/Sequencing_with_the_13k_mutants/for_analysis/all_fastqs_available/TNSeq_barseq_outputs_using_all_fastqs_combined_37•C_repeated_09_30_2024/poolCount.txt', delimiter='\t')
file2 = pd.read_csv('/usr2/people/shollyt22/shollyt22/TnSeq_BarSeq_sequencings/Sequencing_with_the_13k_mutants/for_analysis/all_fastqs_available/TNSeq_barseq_outputs_using_all_fastqs_combined_06_12_2024/poolCount.txt', delimiter='\t')

# Extract the desired columns from file1
columns_to_add = file1[['KFRB0029_37A1', 'KFRB0029_37B1']]

# Add these columns to file2
file2 = pd.concat([file2, columns_to_add], axis=1)

# Remove the specified columns from file2
columns_to_remove = ['KFRB007_1', 'KFRB008_1', 'KFRB019_1', 'KFRB020_1']
file2 = file2.drop(columns=columns_to_remove)

# Rename columns in file2
columns_to_rename = {
    'KFRB0029_37A1': '37•C (A)',
    'KFRB0029_37B1': '37•C (B)',
    'KFRB021_1': '45•C (A)',
    'KFRB022_1': '45•C (B)',
    'KFRB023_1': '45•C (C)',
    'KFRB024_1': '45•C (D)',
    'KFRB025_1': '50•C (A)',
    'KFRB026_1': '50•C (B)',
    'KFRB027_1': '50•C (C)',
    'KFRB028_1': '50•C (D)'
}
file2 = file2.rename(columns=columns_to_rename)

# Specify the desired column order
desired_column_order = [
    'barcode', 'scaffold', 'pos', 'gene', '37•C (A)', '37•C (B)', 
    '45•C (A)', '45•C (B)', '45•C (C)', '45•C (D)',
    '50•C (A)', '50•C (B)', '50•C (C)', '50•C (D)'
]

# Ensure all desired columns are present in file2
missing_columns = [col for col in desired_column_order if col not in file2.columns]
if missing_columns:
    print(f"Warning: The following columns are missing and cannot be reordered: {missing_columns}")
    # Adjust the desired_column_order to include only available columns
    desired_column_order = [col for col in desired_column_order if col in file2.columns]

# Reorder the columns
file2 = file2[desired_column_order]

# Save the result
file2.to_csv('/usr2/people/shollyt22/shollyt22/TnSeq_BarSeq_sequencings/Sequencing_with_the_13k_mutants/for_analysis/all_fastqs_available/TNSeq_barseq_outputs_using_all_fastqs_combined_37•C_repeated_09_30_2024/poolCount_new_37•C_with_older_50•C_1.txt', index=False, sep='\t')
```
