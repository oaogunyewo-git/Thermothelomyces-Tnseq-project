```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data into a DataFrame
file_path = '/usr2/people/shollyt22/shollyt22/TnSeq_BarSeq_sequencings/Barseq_Temp_study_11_04_2023/TNSeq_barseq_outputs_Temp2_with_mainLocation_01_08_2024/normalized_barseq_pool_counts_01_08_2024_with_mainloc.txt'  # Replace with your actual file path
df = pd.read_csv(file_path, sep='\t')

# Calculate the average of columns OORB0010_1 and OORB0011_1
df['X'] = df[['Normalized_OORB0010_1', 'Normalized_OORB0011_1']].mean(axis=1)

# Calculate the average of columns OORB0012_1 and OORB0013_1
df['Y'] = df[['Normalized_OORB0012_1', 'Normalized_OORB0013_1']].mean(axis=1)

# Scatter plot of X against Y
plt.figure(figsize=(8, 6))
scatter = plt.scatter(df['X'], df['Y'], color='blue', edgecolors='black')
plt.title('Scatter Plot of X against Y')
plt.xlabel('X')
plt.ylabel('Y')

# Annotate with gene values
for i, gene in enumerate(df['gene']):
    plt.annotate(gene, (df['X'][i], df['Y'][i]), textcoords="offset points", xytext=(0,5), ha='center')

plt.grid(True)
plt.show()
```
