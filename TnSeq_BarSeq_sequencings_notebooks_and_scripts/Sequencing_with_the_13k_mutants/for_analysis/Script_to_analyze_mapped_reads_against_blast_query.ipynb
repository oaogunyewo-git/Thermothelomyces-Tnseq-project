### To find total reads that mapped against blast query

```python
def get_mapped_read_names(file_path):
    with open(file_path, 'r') as f:
        # Extract read names from the mapped file
        read_names = [line.split('\t')[0][1:] for line in f if line.startswith('@')]
    return read_names

def get_blastquery_read_names(file_path):
    with open(file_path, 'r') as f:
        # Extract read names with the specified identifier
        read_names = [line.split('@')[1].split()[0] if '@' in line else line.strip('> \n') for line in f if line.startswith('>')]
    return read_names

# File paths
mapped_file_path = "/usr2/people/shollyt22/shollyt22/TnSeq_BarSeq_sequencings/Sequencing_with_the_13k_mutants/for_analysis/Batch_2_sequencing_Apr_26_2024/TNSeq_13k_mutants_mapping_output_without_vector_sequence_jumpstart/TNSeq_13k_mutants_mapping_output_jumpstart_polymerase_Apr_26_2024/Tnseq_13k_mutants_Apr_26_2024_without_vector_seq_jumpstart_mapped.txt"
blastquery_file_path = "/usr2/people/shollyt22/shollyt22/TnSeq_BarSeq_sequencings/Sequencing_with_the_13k_mutants/for_analysis/Batch_2_sequencing_Apr_26_2024/TNSeq_13k_mutants_mapping_output_without_vector_sequence_jumpstart/TNSeq_13k_mutants_mapping_output_jumpstart_polymerase_Apr_26_2024/Tnseq_13k_mutants_Apr_26_2024_without_vector_seq_jumpstart_blastquery.fasta"

# Get read names from the mapped file
mapped_read_names = get_mapped_read_names(mapped_file_path)

# Get read names from the Blastquery file
blastquery_read_names = get_blastquery_read_names(blastquery_file_path)

# Compare read names
common_read_names = set(mapped_read_names) & set(blastquery_read_names)
unmapped_read_names = set(blastquery_read_names) - set(mapped_read_names)

# Save common and unmapped read names to files
common_file_path = "/usr2/people/shollyt22/shollyt22/TnSeq_BarSeq_sequencings/Sequencing_with_the_13k_mutants/for_analysis/Batch_2_sequencing_Apr_26_2024/TNSeq_13k_mutants_mapping_output_without_vector_sequence_jumpstart/TNSeq_13k_mutants_mapping_output_jumpstart_polymerase_Apr_26_2024/common_read_names.txt"
unmapped_file_path = "/usr2/people/shollyt22/shollyt22/TnSeq_BarSeq_sequencings/Sequencing_with_the_13k_mutants/for_analysis/Batch_2_sequencing_Apr_26_2024/TNSeq_13k_mutants_mapping_output_without_vector_sequence_jumpstart/TNSeq_13k_mutants_mapping_output_jumpstart_polymerase_Apr_26_2024/unmapped_read_names.txt"

with open(common_file_path, 'w') as common_file:
    common_file.write('\n'.join(common_read_names))

with open(unmapped_file_path, 'w') as unmapped_file:
    unmapped_file.write('\n'.join(unmapped_read_names))

# Print results
print(f"Total Mapped Read Names: {len(mapped_read_names)}")
print(f"Total Blastquery Read Names: {len(blastquery_read_names)}")
print(f"Number of Common Read Names: {len(common_read_names)}")
print(f"Number of Unmapped Read Names: {len(unmapped_read_names)}")

# Print first 10 common and unmapped read names
print("\nFirst 10 Common Read Names:")
print(list(common_read_names)[:10])

print("\nFirst 10 Unmapped Read Names:")
print(list(unmapped_read_names)[:10])

print(f"Common Read Names saved to {common_file_path}")
print(f"Unmapped Read Names saved to {unmapped_file_path}")
```
