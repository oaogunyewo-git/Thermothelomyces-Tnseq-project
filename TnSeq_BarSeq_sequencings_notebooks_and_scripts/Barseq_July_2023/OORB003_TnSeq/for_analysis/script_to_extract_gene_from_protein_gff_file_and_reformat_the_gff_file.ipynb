```python
import pandas as pd

def split_cell_content(input_file, output_file, column_index, delimiter):
    # Read the input table into a pandas DataFrame without header names
    df = pd.read_csv(input_file, sep='\t', header=None)

    # Convert the column with float64 values to strings
    df[column_index] = df[column_index].astype(str)

    # Split the content of the cell using the delimiter
    df[column_index] = df[column_index].str.split(delimiter)

    # Create new columns for the first and second fragments from the split content
    df['First_Fragment'] = df[column_index].apply(lambda x: x[0] if len(x) > 0 else "")
    df['Second_Fragment'] = df[column_index].apply(lambda x: x[3] if len(x) > 1 else "")

    # Drop the original column containing the split content
    df.drop(column_index, axis=1, inplace=True)
    print(df)
    # Save the resulting DataFrame to the output file as a text file
    df.to_csv(output_file, sep='\t', index=False, header=None)  # Use '\t' as the delimiter for a tab-separated file

if __name__ == "__main__":
    # Example usage:
    input_file = "/usr2/people/shollyt22/shollyt22/Barseq_July_2023/OORB003_TnSeq/for_analysis/extracted_gff_files_from_protein_file.txt"  # Replace with the path to your input file
    output_file = "modified_gff_file_4_annotation_2" 
    column_index = 8  # Replace with the index of the column containing the cell content (0-based index)
    delimiter = ";"  # Replace with the appropriate delimiter used to split the cell content

    split_cell_content(input_file, output_file, column_index, delimiter)
```
