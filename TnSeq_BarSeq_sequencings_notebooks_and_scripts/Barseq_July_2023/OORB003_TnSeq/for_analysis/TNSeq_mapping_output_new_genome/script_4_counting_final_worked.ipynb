```python
# Step 1: Read and Parse the Text File
data_file = '/usr2/people/shollyt22/shollyt22/Barseq_July_2023/OORB003_TnSeq/for_analysis/TNSeq_mapping_output_new_genome/Tnseq_07312023_poolfile'  # Replace with the path to your text file
read_counts = {}

with open(data_file, 'r') as file:
    header = file.readline()  # Read and skip the header row
    for line in file:
        fields = line.strip().split('\t')
        insertion_site = (fields[4], int(fields[6]))
        read_count = int(fields[2])  # Assuming "nTot" field contains the read count
        # You may need to adjust the indices based on the columns' positions in your text file

        # Step 2: Count Reads per Insertion Site
        if insertion_site not in read_counts:
            read_counts[insertion_site] = read_count
        else:
            read_counts[insertion_site] += read_count

# Step 3: Count Reads with Multiple Insertion Sites
reads_with_multiple_sites = sum(1 for count in read_counts.values() if count > 100)

# Step 4: Write Output
output_file = '/usr2/people/shollyt22/shollyt22/Barseq_July_2023/OORB003_TnSeq/for_analysis/TNSeq_mapping_output_new_genome/For_counting/unique_insertions_100reads.txt'  # Replace with your desired output file path

with open(output_file, 'w') as out_file:
    out_file.write(f'Reads with multiple insertion sites: {reads_with_multiple_sites}\n')
```
