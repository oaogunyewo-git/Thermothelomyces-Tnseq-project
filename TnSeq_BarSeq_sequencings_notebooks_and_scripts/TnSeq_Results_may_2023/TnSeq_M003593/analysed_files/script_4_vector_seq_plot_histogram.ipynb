```python
import sys
import csv
import numpy as np
import matplotlib
from matplotlib import pyplot as plt

# To store dataset
chromosome_store = []
position_store = []

file = open('output_with_vector_seq_06_28_2023OORB001_S1_L001_R1_001.fastq_mapped_reads')
lines = file.readlines()

for line in lines:
    if 'vector_seq' in line:
        ret = line.split("\t")
        my_chromosome_number = ret[1]
        position = ret[8]
        chromosome_store.append(my_chromosome_number)
        position_store.append(int(position))

j = 0
positions_on_my_chromosome = []
for i in chromosome_store:
    if i == my_chromosome_number:
        positions_on_my_chromosome.append(position_store[j])
    j = j + 1

plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True

# Plot the histogram
plt.hist(positions_on_my_chromosome, bins=50, color='blue', alpha=0.7)
plt.xlabel("Position on Chromosome")
plt.ylabel("Frequency")
plt.title("Histogram of Mapped Reads on Chromosome " + my_chromosome_number)
plt.savefig('vector_seq_1_histogram.png')
plt.show()
```
