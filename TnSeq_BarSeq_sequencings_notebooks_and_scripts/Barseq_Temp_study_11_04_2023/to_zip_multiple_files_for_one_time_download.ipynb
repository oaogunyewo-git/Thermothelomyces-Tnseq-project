```python
import os
import zipfile

folder = "/usr2/people/shollyt22/shollyt22/TnSeq_BarSeq_sequencings/Barseq_Temp_study_11_04_2023"
zip_name = "Barseq_Temp_study_11_04_2023_barseq_notebooks_and_scripts.zip"

with zipfile.ZipFile(zip_name, "w") as z:
    for root, dirs, files in os.walk(folder):
        for f in files:
            if f.endswith(".py") or f.endswith(".ipynb"):
                full_path = os.path.join(root, f)
                # Write file keeping folder structure inside zip
                z.write(full_path, os.path.relpath(full_path, folder))

print("Done! Created:", zip_name)
```
