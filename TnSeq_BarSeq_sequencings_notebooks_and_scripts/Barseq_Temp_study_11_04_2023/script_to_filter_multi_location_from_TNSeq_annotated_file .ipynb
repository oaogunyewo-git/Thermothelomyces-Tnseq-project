## To count the total number of barcodes per scaffold position

```python
import pandas as pd

def count_total_barcode_per_scaffold(tsv_file):
    # Read the TSV file into a DataFrame
    df = pd.read_csv(tsv_file, sep='\t')

    # Group by 'scaffold' and count the number of unique 'barcode's in each group
    total_barcode_per_scaffold = df.groupby('scaffold')['barcode'].nunique()

    return total_barcode_per_scaffold

# Example usage
tsv_file = '/usr2/people/shollyt22/shollyt22/TnSeq_BarSeq_sequencings/Barseq_Temp_study_11_04_2023/TNseq_single_position_nMainLocation_annotated_file.tsv'
total_barcode_per_scaffold = count_total_barcode_per_scaffold(tsv_file)
print(total_barcode_per_scaffold)
```
