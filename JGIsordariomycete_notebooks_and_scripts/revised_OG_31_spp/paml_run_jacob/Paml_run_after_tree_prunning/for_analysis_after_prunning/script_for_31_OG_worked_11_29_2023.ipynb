```python
#load required packages
import sys
import csv
#import pandas as pd
from sys import argv
import os
from scipy.stats import chi2
import numpy as np
# Store all dataset
all_store_data = {}

# To store the orthogroup names
orthogroup_name = []

# Open the output file for writing
with open('final_output_after_pruning_11_29_2023.csv', 'w', newline='') as f:
    writer = csv.writer(f)

    # Store the list of all single-copy orthogroup names
    single_copy_ortho = []
    with open('/usr2/people/shollyt22/shollyt22/JGIsordariomycete/revised_OG_31_spp/paml_run/for_analysis/singlecopyorthogroups.txt') as proj2file:
        proj2reader = csv.reader(proj2file, delimiter=' ')
        for line in proj2reader:
            single_copy_ortho.append(line[0])

    # Loop over each orthogroup
    for myorthogroup_name in single_copy_ortho:
        my_outfile = '/usr2/people/shollyt22/shollyt22/JGIsordariomycete/revised_OG_31_spp/paml_run_jacob/Paml_run_after_tree_prunning/output_files_alt/' + myorthogroup_name + '.altout'
        try:
            file = open(my_outfile)
        except:
            continue

        lines = file.readlines()
        my_lnL_alt = 0  # Initialize my_lnL_alt with a default value
        my_alt_omega_a = 0  # Initialize my_alt_omega_a with a default value
        my_alt_omega_b = 0  # Initialize my_alt_omega_b with a default value

        for line in lines:
            if 'lnL' in line:
                ret = line.split(":")
                my_lnL_alt_splitA = ret[3]
                ret = my_lnL_alt_splitA.split("+")
                my_lnL_alt = ret[0]

            if '(dN/dS)' in line:
                ret = line.split(":")
                my_alt_omega = ret[1]
                ret = my_alt_omega.split(" ")
                my_alt_omega_a = ret[2]
                my_alt_omega_bb = ret[3]
                ret = my_alt_omega_bb.split("\n")
                my_alt_omega_b = ret[0]

        my_outfile = '/usr2/people/shollyt22/shollyt22/JGIsordariomycete/revised_OG_31_spp/paml_run_jacob/Paml_run_after_tree_prunning/output_files_nul/' + myorthogroup_name + '.nulout'
        file = open(my_outfile)
        lines = file.readlines()
        my_lnL_nul = 0  # Initialize my_lnL_nul with a default value

        for line in lines:
            if 'lnL' in line:
                ret = line.split(":")
                my_lnL_nul_splitA = ret[3]
                ret = my_lnL_nul_splitA.split("+")
                my_lnL_nul = ret[0]

            if '(dN/dS)' in line:
                ret = line.split("=")
                my_nul_omega = ret[1]

        my_diff_lnL = int(float(my_lnL_alt)) - int(float(my_lnL_nul))
        my_diff_dof = 1
        chisq = chi2.sf(2 * my_diff_lnL, my_diff_dof)
        my_final_list = [myorthogroup_name, my_alt_omega_a, my_alt_omega_b, chisq]

        # Print the data to the console
        print(my_final_list)

        # Write the data to the CSV file
        writer.writerow(my_final_list)

print("Data saved successfully.")
```
