```python
import csv
from statsmodels.stats import multitest

# Read orthogroups and p-values from a CSV file
input_filename = '/usr2/people/shollyt22/shollyt22/JGIsordariomycete/revised_OG_31_spp/paml_run_jacob/paml_run_after_MACSE_alignment/for_analysis/final_output_p_value_only.csv'
output_filename = '/usr2/people/shollyt22/shollyt22/JGIsordariomycete/revised_OG_31_spp/paml_run_jacob/paml_run_after_MACSE_alignment/for_analysis/final_output_adjusted_p_values_12_23_2023_filtered.csv'
print(input_filename)
orthogroups = []
p_values = []

with open(input_filename, 'r') as csvfile:
    reader = csv.reader(csvfile)
    header = next(reader)  # Read the header row
    for row in reader:
        orthogroup, p_value = row
        orthogroups.append(orthogroup)
        p_values.append(float(p_value))

# Calculate adjusted p-values using the Benjamini-Hochberg (FDR) correction
adjusted_p_values = multitest.fdrcorrection(p_values)[1]

# Save the orthogroups, original p-values, and their corresponding adjusted p-values to a CSV file
with open(output_filename, 'w', newline='') as outfile:
    writer = csv.writer(outfile)
    writer.writerow(['Orthogroup', 'Original P-value', 'Adjusted P-value'])
    for orthogroup, orig_p, adj_p in zip(orthogroups, p_values, adjusted_p_values):
        writer.writerow([orthogroup, orig_p, adj_p])

print(f"Adjusted p-values saved to {output_filename}")
```

```python
import csv
from statsmodels.stats import multitest

# Read orthogroups and p-values from a CSV file
input_filename = '/usr2/people/shollyt22/shollyt22/JGIsordariomycete/revised_OG_31_spp/paml_run_jacob/paml_run_after_MACSE_alignment/for_analysis/final_output_p_value_only.csv'
output_filename = '/usr2/people/shollyt22/shollyt22/JGIsordariomycete/revised_OG_31_spp/paml_run_jacob/paml_run_after_MACSE_alignment/for_analysis/final_output_adjusted_p_values_12_23_2023_filtered.csv'
print(input_filename)
orthogroups = []
p_values = []

with open(input_filename, 'r') as csvfile:
    reader = csv.reader(csvfile)
    header = next(reader)  # Read the header row
    for row in reader:
        orthogroup, p_value = row
        if p_value:  # Skip rows where p-value is an empty string
            orthogroups.append(orthogroup)
            p_values.append(float(p_value))
        else:
            print(f"Skipping row with empty p-value: {orthogroup}")


# Calculate adjusted p-values using the Benjamini-Hochberg (FDR) correction
adjusted_p_values = multitest.fdrcorrection(p_values)[1]

# Save the orthogroups, original p-values, and their corresponding adjusted p-values to a CSV file
with open(output_filename, 'w', newline='') as outfile:
    writer = csv.writer(outfile)
    writer.writerow(['Orthogroup', 'Original P-value', 'Adjusted P-value'])
    for orthogroup, orig_p, adj_p in zip(orthogroups, p_values, adjusted_p_values):
        writer.writerow([orthogroup, orig_p, adj_p])

print(f"Adjusted p-values saved to {output_filename}")
```
