```python
with open('output_1.txt') as file:
    lines = [line.rstrip() for line in file]
```

```python
import pandas as pd
import csv
```

```python
lines
```

```python
names = []
values = []
for i in lines:
    if "KM" in i:
        names.append(i.split("(")[2].split(')')[0])
    elif "dN/dS" in i:
        values.append(float((i.split("=")[4].split("  dN")[0])))
```

```python
dndsdict = dict(zip(names, values))
```

```python
dndsdf = pd.DataFrame(dndsdict.items())
orthdf = pd.DataFrame(columns = ['','km','kl','','sacc','des'])
with open('/home/jovyan/RNAseq/Sacch annotation mapping/37C_with_annotations.dat', newline = '') as orthfile:                                                                                          
    orthreader = csv.reader(orthfile, delimiter='\t')
    for line in orthreader: 
        orthdf.loc[len(orthdf)] = line
des = dict(zip(orthdf['kl'],orthdf['des']))
dndsdf['description'] = dndsdf[0].map(des)
dndsdf
```

```python
with open('messedupseqs.txt') as file:
    l = [line.rstrip() for line in file]
for i in range(0,len(dndsdf[0])):
    if dndsdf[0][i] in l:
        dndsdf = dndsdf.drop([i])
dndsdf
```

```python
dndsdf = dndsdf.sort_values(1,ascending=False)
dndsdf
```

```python
dndsdf.to_csv('/home/jovyan/PAMLoutput/dNdS_KMKL_sorted_1.csv', index=False)
```
