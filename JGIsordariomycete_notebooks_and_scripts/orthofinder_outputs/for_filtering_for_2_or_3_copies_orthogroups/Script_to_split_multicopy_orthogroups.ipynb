# To split orthogroups with 2 or 3 copies into individual copies

```python
import pandas as pd

# Load the Orthogroups.tsv file
df = pd.read_csv('/usr2/people/shollyt22/shollyt22/JGIsordariomycete/orthofinder_outputs/Orthogroups.tsv', sep='\t')

# Filter for orthologs with 2 or 3 copies
filtered_df = df[df.apply(lambda row: 2 <= sum(row.notnull()) <= 3, axis=1)]

# Iterate through the filtered orthogroups and create separate files for each copy
for index, row in filtered_df.iterrows():
    orthogroup_name = row['Orthogroup']
    for species, gene_id in row.dropna().items():
        # Create a new file for each orthologous copy
        file_name = f'{orthogroup_name}_{species}.fasta'
        with open(file_name, 'w') as file:
            # Write the gene sequence to the file
            file.write(f'>{orthogroup_name}_{species}\n{gene_id}\n')

print("Files created successfully.")
```
