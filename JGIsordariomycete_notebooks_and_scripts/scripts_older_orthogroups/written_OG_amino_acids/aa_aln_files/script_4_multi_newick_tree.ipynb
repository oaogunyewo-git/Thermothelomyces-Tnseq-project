```python
from Bio import Phylo, AlignIO
from Bio.Phylo.TreeConstruction import DistanceCalculator, DistanceTreeConstructor
import os
import matplotlib.pyplot as plt
```

```python
alignment_dir = "/usr2/people/shollyt22/shollyt22/JGIsordariomycete/written_OG_amino_acids/aa_aln_files/"
output_dir = "/usr2/people/shollyt22/shollyt22/JGIsordariomycete/written_OG_amino_acids/aa_aln_files/aln_phylo_outfiles/"
print(alignment_dir)
```

```python
alignment_files = os.listdir(alignment_dir)
alignment_files = [file for file in alignment_files if file.endswith(".fasta")]
print(alignment_files)
print(f"Number of alignment files: {len(alignment_files)}")
```

```python
for index, alignment_file in enumerate(alignment_files):
    alignment_path = os.path.join(alignment_dir, alignment_file)
    alignment = AlignIO.read(alignment_path, "clustal")
    #print(alignment)
    calculator = DistanceCalculator("identity")
    distance_matrix = calculator.get_distance(alignment)
    constructor = DistanceTreeConstructor()
    tree = constructor.upgma(distance_matrix)
    alignment_name = os.path.splitext(alignment_file)[0]
    tree_file = os.path.join(output_dir, f"tree_{alignment_name}.nwk")
    Phylo.write(tree, tree_file, "newick")

    print(f"Processed alignment file {index + 1}/{len(alignment_files)}: {alignment_file}")

print("Processing completed!")
```
