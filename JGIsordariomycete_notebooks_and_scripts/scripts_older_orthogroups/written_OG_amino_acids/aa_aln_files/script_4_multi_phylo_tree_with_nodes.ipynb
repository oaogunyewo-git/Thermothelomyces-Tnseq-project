```python
from Bio import Phylo
import os
import matplotlib.pyplot as plt

tree_dir = "/usr2/people/shollyt22/shollyt22/JGIsordariomycete/written_OG_amino_acids/aa_aln_files/aln_phylo_outfiles/newick_trees_no_nodes/"
output_dir = "/usr2/people/shollyt22/shollyt22/JGIsordariomycete/written_OG_amino_acids/aa_aln_files/aln_phylo_outfiles/tree_plots/treeplots_with_no nodes/"

tree_files = [file for file in os.listdir(tree_dir) if file.endswith(".nwk")]

for index, tree_file in enumerate(tree_files):
    tree_path = os.path.join(tree_dir, tree_file)

    # Read the tree file
    tree = Phylo.read(tree_path, "newick")

    # Determine the output file path
    output_file = os.path.join(output_dir, os.path.splitext(tree_file)[0] + ".png")

    # Plot and save the tree
    Phylo.draw(tree, do_show=False)
    Phylo.draw(tree, do_show=False)
    Phylo.draw(tree, do_show=False, show_confidence=False)
    Phylo.draw(tree, do_show=False, branch_labels=lambda c: c.branch_length)

    # Save the plot
    Phylo.draw(tree, do_show=False, branch_labels=lambda c: c.branch_length)
    Phylo.draw(tree, do_show=False, branch_labels=lambda c: c.branch_length)
    Phylo.draw(tree, do_show=False, branch_labels=lambda c: c.branch_length)

    # Save the plot
    Phylo.draw(tree, do_show=False)
    plt.savefig(output_file)

    print(f"Processed tree file {index + 1}/{len(tree_files)}: {tree_file}")

print("Processing completed!")
```
