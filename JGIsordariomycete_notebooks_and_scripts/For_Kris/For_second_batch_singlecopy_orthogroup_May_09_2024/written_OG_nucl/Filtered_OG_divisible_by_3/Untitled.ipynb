```python
import os

def rename_files(directory, old_extension, new_suffix, new_extension):
    """
    Rename files in the directory by appending a new suffix and changing the extension.
    """
    for filename in os.listdir(directory):
        if filename.endswith(old_extension):
            base_id = filename.split(old_extension)[0]
            new_filename = f"{base_id}_{new_suffix}.{new_extension}"
            old_file_path = os.path.join(directory, filename)
            new_file_path = os.path.join(directory, new_filename)
            os.rename(old_file_path, new_file_path)
            print(f"Renamed {filename} to {new_filename}")

# Define the second directory
directory_two = '/usr2/people/shollyt22/shollyt22/JGIsordariomycete/For_second_batch_singlecopy_orthogroup_May_09_2024/written_OG_nucl/Filtered_OG_divisible_by_3/muscle_realignment/alignment_msf/copy_of_files_to_be_renamed'

# Rename files in the second directory
rename_files(directory_two, 'msf', 'aligned', 'phylip')
```
