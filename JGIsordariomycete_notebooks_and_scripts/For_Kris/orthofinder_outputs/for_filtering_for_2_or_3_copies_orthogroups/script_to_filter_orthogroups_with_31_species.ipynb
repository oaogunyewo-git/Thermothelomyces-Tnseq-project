```python
import pandas as pd

# Load the original Orthogroups.tsv file
original_orthogroups_path = '/usr2/people/shollyt22/shollyt22/JGIsordariomycete/orthofinder_outputs/for_filtering_for_2_or_3_copies_orthogroups/filtered_Orthogroups_31_species.tsv'
df = pd.read_csv(original_orthogroups_path, sep='\t')

# Define a list of species of interest
species_of_interest = [
    'podbul-1', 'achmac-1', 'arnari-1', 'podbac-1', 'podset-1', 'podcom-1', 'podans-001',
    'podbel-1', 'podpse-1', 'canare-1', 'madmyc-1', 'chathe-1', 'triant-1',
    'theter-1', 'achstr-1', 'stalon-1', 'mycthe-1', 'parapp-1', 'parhyr-1',
    'chalep-1', 'dicfun-1', 'corina-1', 'chacoc-1', 'chafim-1', 'cornov-1',
    'corsep-1', 'thethe-1', 'thehet-1', 'crahot-1', 'crathe-1', 'neucra-03'
]

# Function to count copies for each species in a row
def count_species_copies(row, species):
    if species in row and pd.notna(row[species]):
        return len(row[species].split(','))  # Split the cell content by commas and count
    else:
        return 0  # If the species is not in the row or is NaN, return zero

# Create a new DataFrame to store the results
results = pd.DataFrame()

# Add the Orthogroup column to the results DataFrame
results['Orthogroup'] = df['Orthogroup']

# Iterate over the list of species and calculate the number of copies for each
for species in species_of_interest:
    column_name = f'Copies_in_{species}'
    results[column_name] = df.apply(lambda row: count_species_copies(row, species), axis=1)

# Display the number of copies for each orthogroup for all species of interest
print("Number of copies for each orthogroup in the specified species:")
print(results)

# Optionally, save the results to a new TSV file
output_path = '/usr2/people/shollyt22/shollyt22/JGIsordariomycete/orthofinder_outputs/for_filtering_for_2_or_3_copies_orthogroups/filtered_orthogroups_copies_summary.tsv'
results.to_csv(output_path, sep='\t', index=False)

print(f"Summary of orthogroups with copies count for each species saved to {output_path}")
```

### find orthogroups where every specified species has a non-empty gene identifier (no missing data), you can focus on rows in which there are no empty or null values for any species. This typically involves checking that each cell (after excluding the Orthogroup column) has a value

```python
import pandas as pd

# Load the original Orthogroups.tsv file
original_orthogroups_path = '/usr2/people/shollyt22/shollyt22/JGIsordariomycete/orthofinder_outputs/for_filtering_for_2_or_3_copies_orthogroups/filtered_Orthogroups_31_species.tsv'
df = pd.read_csv(original_orthogroups_path, sep='\t')

# Function to check if every species has a gene identifier (non-empty)
def all_species_have_gene_identifier(row):
    # Check if every cell (excluding 'Orthogroup') is not NaN or empty
    for value in row[1:]:  # Skip the 'Orthogroup' column
        if pd.isna(value) or value.strip() == '':  # If any cell is empty or NaN
            return False  # Disqualify this row
    return True  # If all cells are non-empty, the row is valid

# Filter orthogroups where every species has a non-empty gene identifier
complete_orthogroups = df[df.apply(all_species_have_gene_identifier, axis=1)]

# Save the filtered DataFrame to a new TSV file
output_path = '/usr2/people/shollyt22/shollyt22/JGIsordariomycete/orthofinder_outputs/for_filtering_for_2_or_3_copies_orthogroups/filtered_orthogroups_complete_no_empty_gene_identifier.tsv'
complete_orthogroups.to_csv(output_path, sep='\t', index=False)

print(f"Orthogroups where all species have a gene identifier saved to {output_path}")
```

```python
import pandas as pd

# Load the orthogroups copies summary file
summary_path = '/usr2/people/shollyt22/shollyt22/JGIsordariomycete/orthofinder_outputs/for_filtering_for_2_or_3_copies_orthogroups/filtered_orthogroups_copies_summary.tsv'
df = pd.read_csv(summary_path, sep='\t')

# Function to check if an orthogroup has only one copy for all species
def has_only_one_copy_per_species(row):
    # Check if each value is exactly 1
    for value in row[1:]:  # Skip the 'Orthogroup' column
        if value != 1:
            return False  # If any value is not 1, return False
    return True  # Return True if all values are 1

# Filter the DataFrame for orthogroups with one copy per species
single_copy_df = df[df.apply(has_only_one_copy_per_species, axis=1)]

# Save the filtered DataFrame to a new TSV file
output_path = '/usr2/people/shollyt22/shollyt22/JGIsordariomycete/orthofinder_outputs/for_filtering_for_2_or_3_copies_orthogroups/filtered_orthogroups_with_one_copy_per_species.tsv'
single_copy_df.to_csv(output_path, sep='\t', index=False)

print(f"Orthogroups with one copy per species saved to {output_path}")
```

```python
import pandas as pd

# Load the original Orthogroups.tsv file
original_orthogroups_path = '/usr2/people/shollyt22/shollyt22/JGIsordariomycete/orthofinder_outputs/for_filtering_for_2_or_3_copies_orthogroups/filtered_orthogroups_complete_no_empty_gene_identifier.tsv'
df = pd.read_csv(original_orthogroups_path, sep='\t')

# Function to check if every species in a row has only one copy (no commas)
def all_species_have_one_copy(row):
    # Iterate over each species (excluding 'Orthogroup')
    for value in row[1:]:  # Skip the 'Orthogroup' column
        if pd.notna(value):  # Ensure it's not NaN
            if ',' in value:  # If there are multiple copies (comma-separated)
                return False  # Disqualify the orthogroup
    return True  # Return True if all species have only one copy (no commas)

# Filter orthogroups where each species has exactly one copy
single_copy_orthogroups = df[df.apply(all_species_have_one_copy, axis=1)]
total_count_single_copy_orthogroups = len(single_copy_orthogroups)
# Save the filtered DataFrame to a new TSV file
output_path = '/usr2/people/shollyt22/shollyt22/JGIsordariomycete/orthofinder_outputs/for_filtering_for_2_or_3_copies_orthogroups/filtered_orthogroups_single_copy_per_species_original_2.tsv'
single_copy_orthogroups.to_csv(output_path, sep='\t', index=False)

print(f"Orthogroups with a single copy in all species saved to {output_path}")
print("total number of orthogroups generated:", total_count_single_copy_orthogroups)
```
