```python
import os
from Bio import AlignIO
from Bio.SeqRecord import SeqRecord

# Define the directory containing your Clustal alignment files
input_directory = "/usr2/people/shollyt22/shollyt22/JGIsordariomycete/For_second_batch_singlecopy_orthogroup_May_09_2024/written_OG_nucl/Filtered_OG_divisible_by_3/amino_acid_alignments/alignment_aln_AAs"
output_directory = "/usr2/people/shollyt22/shollyt22/JGIsordariomycete/For_second_batch_singlecopy_orthogroup_May_09_2024/written_OG_nucl/Filtered_OG_divisible_by_3/amino_acid_alignments/alignment_aln_AAs/reordered"

# Ensure the output directory exists
os.makedirs(output_directory, exist_ok=True)

# Define the desired order of species (sequence IDs)
desired_order = [
    "crathe", "crahot", "thehet", "thethe", "corsep", "cornov", 
    "chafim", "chacoc", "corina", "dicfun", "chalep", "parhyr", 
    "parapp", "mycthe", "stalon", "achstr", "achmac", "theter", 
    "triant", "chathe", "madmyc", "canare", "podpse", "probel", 
    "podans", "podcom", "podset", "podbac", "arnari", "podbul", 
    "neucra"
]

# Get a list of Clustal alignment files from the input directory
alignment_files = [f for f in os.listdir(input_directory) if f.endswith('.aln')]

# Function to reorder a single Clustal alignment file
def reorder_alignment(input_file, output_file, desired_order):
    # Read the Clustal alignment
    alignment = AlignIO.read(input_file, "clustal")
    
    # Create a dictionary of sequences for easy reordering
    seq_dict = {record.id: record for record in alignment}
    
    # Reorder the sequences according to the desired order
    reordered_records = [seq_dict[species] for species in desired_order if species in seq_dict]
    
    # Create a new alignment with the reordered sequences
    reordered_alignment = AlignIO.MultipleSeqAlignment(reordered_records)
    
    # Write the reordered alignment back to a Clustal file
    AlignIO.write(reordered_alignment, output_file, "clustal")

# Iterate over each Clustal alignment file and reorder the sequences
for filename in alignment_files:
    input_file = os.path.join(input_directory, filename)
    output_file = os.path.join(output_directory, f"reordered_{filename}")
    
    try:
        reorder_alignment(input_file, output_file, desired_order)
        print(f"Reordered alignment saved to {output_file}")
    except Exception as e:
        print(f"Error processing file {input_file}: {e}")

print("Reordering process completed for all files.")
```
