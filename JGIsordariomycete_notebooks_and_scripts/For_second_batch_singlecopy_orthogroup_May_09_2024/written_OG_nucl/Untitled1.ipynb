```python
from Bio import AlignIO
from Bio.SeqRecord import SeqRecord

# Define the path to your Clustal alignment file
clustal_file = "/usr2/people/shollyt22/shollyt22/JGIsordariomycete/For_second_batch_singlecopy_orthogroup_May_09_2024/written_OG_nucl/Filtered_OG_divisible_by_3/amino_acid_alignments/alignment_aln_AAs/OG0001636.aln"

# Define the desired order of species (sequence IDs)
desired_order = [
    "crathe",
    "crahot",
    "thehet", 
    "thethe", 
    "corsep", 
    "cornov", 
    "chafim",
    "chacoc", 
    "corina", 
    "dicfun", 
    "chalep",
    "parhyr", 
    "parapp", 
    "mycthe", 
    "stalon",
    "achstr", 
    "achmac", 
    "theter", 
    "triant", 
    "chathe",
    "madmyc", 
    "canare", 
    "podpse", 
    "probel",
    "podans", 
    "podcom", 
    "podset", 
    "podbac",
    "arnari", 
    "podbul", 
    "neucra", 
]

# Read the Clustal alignment
alignment = AlignIO.read(clustal_file, "clustal")

# Create a dictionary of sequences for easy reordering
seq_dict = {record.id: record for record in alignment}

# Reorder the sequences according to the desired order
reordered_records = [seq_dict[species] for species in desired_order if species in seq_dict]

# Create a new alignment with the reordered sequences
reordered_alignment = AlignIO.MultipleSeqAlignment(reordered_records)

# Write the reordered alignment back to a Clustal file
output_file = "/usr2/people/shollyt22/shollyt22/JGIsordariomycete/For_second_batch_singlecopy_orthogroup_May_09_2024/written_OG_nucl/Filtered_OG_divisible_by_3/amino_acid_alignments/alignment_aln_AAs/reordered/reordered_OG0001636.aln"
AlignIO.write(reordered_alignment, output_file, "clustal")

print(f"Reordered alignment saved to {output_file}")
```
